## Hello, World! — Minimal example

The simplest way to print text to the screen in Python is to use the `print()` function.

Run the code cell above to see the output.

### Explanation — step by step

- `print()` function: prints text to the screen (terminal or notebook output).
- String literal: `\
` is text enclosed in quotes. You can use single (`'`) or double (`\"`) quotes.
- Parentheses and argument: the text inside `print()` is the argument passed to the function.
- Newline behavior: `print()` adds a newline after the output by default.
- Syntax note: in Python 3 you must use parentheses with `print`, e.g. `print(\"text\")`.

## Expanded examples

Below are small examples that show variables, functions, and f-strings (string interpolation).
Run the code cell that follows to see each example in action.

```python
# Store text in a variable and print it
message = "Hello, world!"
print(message)

# Define a function to reuse code
def say_hello(name="World"):
    print(f"Hello, {name}!")

# Call the function with and without an argument
say_hello()
say_hello("Alice")
```
