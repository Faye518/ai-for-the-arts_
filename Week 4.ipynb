```python
from collections import Counter
poem = "Shall I compare thee to a summer's day? Thou art more lovely and more temperate: Rough winds do shake the darling buds of May, And summerâ€™s lease hath all too short a date:"
words = [w.lower() for w in poem.split()]
counts = Counter(words)
most_common = counts.most_common(1)[0]
print ("Most frequent word:", most_common[0], "appears", most_common[1], "times")
```

```python
# Import the webbrowser module to open URLs in a web browser
import webbrowser
# Import the requests library to make HTTP requests
import requests
# Print a prompt asking the user to find an old website
print("Shall we hunt down an old website?")
# Get the website URL from user input
site = input("Type a website URL: ")
# Get the date/time from user input (format: YYYYMMDD)
era = input("Type year, month, and date, e.g., 20150613: ")
# Construct the URL to query the Wayback Machine API
url = "http://archive.org/wayback/available?url=%s&timestamp=%s" % (site, era)
# Send an HTTP GET request to the Wayback Machine API
response = requests.get(url)
# Parse the response as JSON data
data = response.json()
# Try to extract and open the archived website URL
try:
    # Navigate through the JSON data to get the closest archived snapshot URL
    old_site = data["archived_snapshots"]["closest"]["url"]
    # Print the found archived URL
    print("Found this copy: ", old_site)
    # Inform the user that the browser will open
    print("It should appear in your browser.")
    # Open the archived website in the default web browser
    webbrowser.open(old_site)
# Handle the case where the archived website could not be found
except:
    # Print an error message if the website was not found in the archive
    print("Sorry, could not find the site.")
```
